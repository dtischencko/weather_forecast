```python
import os
import glob
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pandas import Series, DataFrame
```

```python
path = r'../data/'
all_files = glob.glob(os.path.join(path, "*.csv"))
sup_file = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
df = DataFrame(sup_file)
df
```

```python
df.describe()
```

```python
df.hist(figsize=(20, 15), bins=50)
plt.show()
```

### Performing raw dataset
Using only numeric columns

```python
df_num = df.drop(['name', 'conditions', 'datetime', 'description', 'moonphase', 'precipprob', 'preciptype', 'snow', 'snowdepth', 'stations', 'sunrise', 'sunset', 'severerisk'], axis=1)
df_num['windgust'] = df_num['windgust'].fillna(0.0)
df_num
```

```python
print(df_num.isna().any(), '\n\n\n', df_num.isnull().any())
```

### Drop na-values

```python
df_num.loc[df_num['sealevelpressure'].isna()]
```

```python
df_num = df_num.dropna(axis=0)
df_num
```

### Rename target column in dataset

```python
df_num = df_num.rename(columns={'icon': 'target'})
df_num
```

```python
df_num.to_csv("../data/FULL_DATA_SET")
```
